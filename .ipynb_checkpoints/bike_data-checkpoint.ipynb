```python
# Import pandas 
import pandas as pd
import glob
from glob import iglob
import os
from os import listdir
```

```python
print("Current Working Directory " , os.getcwd())
```

```python
# reading and combining all csv files  
df_bike = pd.concat(map(pd.read_csv, glob.glob('/Users/gastonalvarado/Desktop/cwbootcamp/tableau_assignment/bike_data_files/*.csv')))

df_bike.rename(columns={'Trip Duration': 'trip_duration', 'Start Time': 'start_time', 'Stop Time': 'stop_time', 
                   'Start Station ID': 'start_station','Start Station Name': 'start_station_name',
                   'Start Station Latitude': 'start_station_latitude', ' Start Station Longitude': 'start_station_longitude', 
                   'End Station ID': 'end_station_id', 'End Station Name': 'end_station_name',
                   'End Station Latitude': 'end_ststion_latitude', 'End Station Longitude': 'end_station_longitude', 
                   'Bike ID': 'bike_id', 'User Type': 'user_type', 'Birth Year': 'birth_year', 'Gender': 'gender'}, inplace=True)

df_bike.head()
```

```python
df_bike['start_time'] = pd.to_datetime(df_bike['start_time'], errors='coerce')
df_bike['stop_time'] = pd.to_datetime(df_bike['stop_time'], errors='coerce')
df_bike.head()
```

```python
# Count rows and columns 
df_bike.shape
```

```python
# Check for missing values 
df_bike.count()
```

```python
df2=df_bike[df_bike.start_time.notnull()]
df3=df2[df2.stop_time.notnull()]
df_bike_clean=df3[df3.birth_year.notnull()]
df_bike_clean.head()
```

```python
# Check for missing values 
df_bike_clean.count()
```

```python
df_bike_clean.dtypes
```

```python
# We see take the Checkout Date and extract the Month, Year, and Day of the Week 
df_bike_clean['start_time_clean'] = pd.to_datetime(df_bike_clean['start_time'])
df_bike_clean['stop_time_clean'] = pd.to_datetime(df_bike_clean['stop_time'])
df_bike_clean['year'] = df_bike_clean['start_time'].dt.year
df_bike_clean['month'] = df_bike_clean['start_time'].dt.month
df_bike_clean['trip_date'] = df_bike_clean['start_time'].dt.day
df_bike_clean['trip_day_of_week'] = df_bike_clean['start_time'].dt.weekday_name
```

```python
df_bike_clean.head()
```

```python
# Convert bicycle ID from object to integer
df_bike_clean2 = df_bike_clean["bike_id"].astype(int)
#df_bike_clean2=df_bike_clean["bike_id"] 
df_bike_clean2.head()
```
