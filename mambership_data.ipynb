```python
jan_mar_2017 = pd.read_csv('./jan-mar-2017.csv', index_col=0)
apr_jun_2017 = pd.read_csv('./apr-jun-2017.csv', index_col=0)
jul_sep_2017 = pd.read_csv('./jul-sep-2017.csv', index_col=0)
oct_dec_2017 = pd.read_csv('./onc-dec-2017.csv', index_col=0)
jan_mar_2018 = pd.read_csv('./jan-mar-2018.csv', index_col=0)
apr_jun_2018 = pd.read_csv('./apr-jun-2018.csv', index_col=0)
jul_sep_2018 = pd.read_csv('./jul-sep-2018.csv', index_col=0)
oct_dec_2018 = pd.read_csv('./oct-dec-2018.csv', index_col=0)
jan_mar_2019 = pd.read_csv('./jan-mar-2019.csv', index_col=0)
apr_jun_2019 = pd.read_csv('./apr-jun-2019.csv', index_col=0)
```

```python
nj_membership = pd.concat([jan_mar_2017, apr_jun_2017,jul_sep_2017,oct_dec_2017,jan_mar_2018,apr_jun_2018,jul_sep_2018,oct_dec_2018,jan_mar_2019,apr_jun_2019]).reset_index()
nj_membership.head()
```

```python
nj_membership['Date'] = pd.to_datetime(nj_membership['Date'], errors='coerce')
nj_membership
```

```python
# Count rows and columns 
nj_membership.shape
```

```python
# Check for missing values 
nj_membership.count()
```

```python
# We see take the Date and extract the Month, Year, and Day of the Week
nj_membership['year'] = nj_membership['Date'].dt.year
nj_membership['month'] = nj_membership['Date'].dt.month
nj_membership['trip_date'] = nj_membership['Date'].dt.day
nj_membership['trip_day_of_week'] = nj_membership['Date'].dt.weekday_name
```

```python
nj_membership.to_csv('nj_membership_data_2017_2019.csv', encoding='utf-8', index=False)
```
